```python
#Display the first image in train images
def img_show(img):
    plt.imshow(img, cmap='gray')
    plt.axis('off')  # Hide axes
    plt.show()
```

```python
img_path1 = "C:/Users/jcasu/OneDrive/Desktop/img_smile.jpg"
img_path2 = "C:/Users/jcasu/OneDrive/Desktop/Smiling_Lady.jpg"
img1 = Image.open(img_path1).convert('L')  # Convert to grayscale
img2 = Image.open(img_path2).convert('L')
img_show(img1)
img_show(img2)
```

```python
img1 = img1.resize((48, 48))  # Resize to uniform size (e.g., 48x48)
img2 = img2.resize((48, 48))
img_show(img1)
img_show(img2)
```
